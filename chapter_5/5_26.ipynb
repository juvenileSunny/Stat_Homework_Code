```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Degrees of freedom
df = 8
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

# Critical t-values
t_010 = t.ppf(0.90, df)
t_005 = t.ppf(0.95, df)
t_001 = t.ppf(0.99, df)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Plot 1: P(-t(8,.10) <= T <= t(8,.10)) => Middle 80%
axes[0].plot(x, y, label='t(8) PDF')
axes[0].fill_between(x, y, where=(x >= -t_010) & (x <= t_010), color='skyblue')
axes[0].set_title("P(-t(8,0.10) ≤ T ≤ t(8,0.10)) ≈ 0.80")

# Plot 2: P(-t(8,.05) <= T <= t(8,.01)) => 0.94 approx
axes[1].plot(x, y, label='t(8) PDF')
axes[1].fill_between(x, y, where=(x >= -t_005) & (x <= t_001), color='lightgreen')
axes[1].set_title("P(-t(8,0.05) ≤ T ≤ t(8,0.01)) ≈ 0.94")

# Plot 3: P(t(8,.05) <= T <= t(8,.01)) => 0.04
axes[2].plot(x, y, label='t(8) PDF')
axes[2].fill_between(x, y, where=(x >= t_005) & (x <= t_001), color='salmon')
axes[2].set_title("P(t(8,0.05) ≤ T ≤ t(8,0.01)) ≈ 0.04")

# Plot 4: P(T > t(8,.05)) => 0.05
axes[3].plot(x, y, label='t(8) PDF')
axes[3].fill_between(x, y, where=(x >= t_005), color='orange')
axes[3].set_title("P(T > t(8,0.05)) ≈ 0.05")

# Formatting
for ax in axes:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.set_xlim([-4, 4])
    ax.set_xlabel('t')
    ax.set_ylabel('Density')
    ax.legend()

plt.tight_layout()
plt.show()
```
