```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 14

# x range for plotting
x = np.linspace(0, 40, 500)
y = chi2.pdf(x, df)

# Critical values
a = chi2.ppf(0.05, df)  # P(X > a) = 0.95
b = chi2.ppf(0.95, df)  # P(X > b) = 0.05
c = chi2.ppf(0.99, df)  # P(X < c) = 0.99
d = a  # lower bound of middle 90%
e = b  # upper bound of middle 90%

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, label=r'$\chi^2_{14}$ PDF', color='black')

# Shade P(X > a) = 0.95
x_fill_a = np.linspace(a, 40, 500)
plt.fill_between(x_fill_a, chi2.pdf(x_fill_a, df), color='red', alpha=0.3, label='P(X > a) = 0.95')

# Shade P(X > b) = 0.05
x_fill_b = np.linspace(b, 40, 500)
plt.fill_between(x_fill_b, chi2.pdf(x_fill_b, df), color='blue', alpha=0.3, label='P(X > b) = 0.05')

# Shade P(X < c) = 0.99
x_fill_c = np.linspace(0, c, 500)
plt.fill_between(x_fill_c, chi2.pdf(x_fill_c, df), color='green', alpha=0.3, label='P(X < c) = 0.99')

# Shade middle 90%: P(d < X < e)
x_fill_de = np.linspace(d, e, 500)
plt.fill_between(x_fill_de, chi2.pdf(x_fill_de, df), color='orange', alpha=0.4, label='P(d < X < e) = 0.90')

# Markers for critical values
for val, label in zip([a, b, c, d, e], ['a', 'b', 'c', 'd', 'e']):
    plt.axvline(val, linestyle='--', color='gray')
    plt.text(val, chi2.pdf(val, df)+0.002, f'{label} = {val:.2f}', ha='center', fontsize=10)

plt.title(r'$\chi^2_{14}$ Distribution and Probability Regions')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
