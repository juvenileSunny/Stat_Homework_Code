```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Degrees of freedom
d1, d2 = 8, 12
x = np.linspace(0, 5, 1000)
y = f.pdf(x, d1, d2)

# Critical values
a = f.ppf(0.95, d1, d2)     # P(F > a) = 0.05
b = f.ppf(0.01, d1, d2)     # P(F > b) = 0.99
c = f.ppf(0.90, d1, d2)     # P(F < c) = 0.90
d_val = f.ppf(0.025, d1, d2)
e = f.ppf(0.975, d1, d2)

# Plot the PDF
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='F(8,12) PDF', color='black')

# (1) P(F > a) = 0.05
plt.fill_between(x, y, where=(x > a), color='orange', alpha=0.5, label=f'P(F > {a:.2f}) = 0.05')

# (2) P(F > b) = 0.99
plt.fill_between(x, y, where=(x > b), color='green', alpha=0.4, label=f'P(F > {b:.2f}) = 0.99')

# (3) P(F < c) = 0.90
plt.fill_between(x, y, where=(x < c), color='blue', alpha=0.3, label=f'P(F < {c:.2f}) = 0.90')

# (4) P(d < F < e) = 0.95
plt.fill_between(x, y, where=((x > d_val) & (x < e)), color='purple', alpha=0.3,
                 label=f'P({d_val:.2f} < F < {e:.2f}) = 0.95')

# Labels and legend
plt.title("Shaded Areas under F(8,12) PDF")
plt.xlabel("F-value")
plt.ylabel("Density")
plt.legend(loc="upper right")
plt.grid(True)
plt.tight_layout()
plt.show()

```
