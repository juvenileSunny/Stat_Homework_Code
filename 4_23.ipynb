```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data: lifetimes in months
lifetimes = np.array([12.5, 14.4, 4.7, 26.6, 18.0, 7.3, 9.8, 36.2, 20.1])

# Sort the data
sorted_lifetimes = np.sort(lifetimes)
n = len(lifetimes)

# Calculate normal scores (quantiles from standard normal distribution)
ranks = np.arange(1, n+1)
normal_scores = stats.norm.ppf((ranks - 0.5) / n)

# Plot normal scores vs sorted lifetimes
plt.figure(figsize=(8, 6))
plt.scatter(normal_scores, sorted_lifetimes, color='blue')
plt.plot(normal_scores, sorted_lifetimes, linestyle='--', color='gray', alpha=0.6)
plt.title("Normal Probability Plot (Q-Q Plot)")
plt.xlabel("Theoretical Quantiles (Normal Scores)")
plt.ylabel("Observed Lifetimes (months)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Output normal scores for reference
list(zip(sorted_lifetimes, np.round(normal_scores, 4)))
```
